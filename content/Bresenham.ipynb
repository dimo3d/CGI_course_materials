```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def bresenham_line(x0, y0, x1, y1):
    # Create an empty list to store the points
    points = []

    # Calculate deltas
    deltaX = abs(x1 - x0)
    deltaY = abs(y1 - y0)
    error = 0

    x, y = x0, y0

    # Determine the direction of iteration
    xStep = 1 if x0 < x1 else -1
    yStep = 1 if y0 < y1 else -1

    if deltaX > deltaY:
        while x != x1:
            points.append((x, y))
            error += deltaY
            if 2 * error >= deltaX:
                y += yStep
                error -= deltaX
            x += xStep
    else:
        while y != y1:
            points.append((x, y))
            error += deltaX
            if 2 * error >= deltaY:
                x += xStep
                error -= deltaY
            y += yStep

    # Don't forget to plot the last point
    points.append((x1, y1))
    return points
```

```python
# Define start and end points
start_point = (1, 2)
end_point = (10, 4)

# Get the points for the line
line_points = bresenham_line(*start_point, *end_point)
```

```python
# Unzip the points into x and y coordinates
x_coords, y_coords = zip(*line_points)

# Plot each point as a square
plt.figure(figsize=(5,5))
plt.scatter(x_coords, y_coords, marker='s', s=100) # 's' for square, s for size
plt.title('Bresenham Line from {} to {}'.format(start_point, end_point))
plt.grid(True)
plt.axis('equal')  # Ensures that squares are shown as squares

# Optional: Set axes limits if you want a specific range
plt.xlim(min(x_coords) - 1, max(x_coords) + 1)
plt.ylim(min(y_coords) - 1, max(y_coords) + 1)

plt.show()
```
